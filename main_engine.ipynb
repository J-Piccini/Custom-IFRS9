```python
def main():
    try:
        ifrs9 = IFRS9Engine(
            portfolio_name = 'Ameriabank',
            reference_snapshot = 202407,
            test = True
        )
        
        for ptf in ifrs9.portfolio_list:
            ifrs9.portfolio_ccf_loop(
                ptf_obj = ptf
            )

            ptf.lel, ptf.tranche_loss_dict, ptf.fmo_loss_bool = ptf.tranching_and_lifetime_expected_loss_calculation(
                tranche_dict = ifrs9.params_application.portfolio_tranches_arr,
                guarantee_dict = ifrs9.params_ccf.ccf_guarantee_dict
                        )
            

    except Exception as e:
        print(f'Error occured: {str(e)}')
        raise

    return ifrs9

if __name__ == '__main__':
    engine = main()
```
